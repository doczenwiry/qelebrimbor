```python
from qelebrimbor.common.attributes_zx import NodeType, EdgeType
from qelebrimbor.volumetric_zx_graph import VolumetricZxGraph
```

```python
vzx = VolumetricZxGraph(
    nodes = [
        (0, NodeType.O),
        (1, NodeType.O),
        (2, NodeType.Z),
        (3, NodeType.X),
        (4, NodeType.O),
        (5, NodeType.O)
    ],
    edges = [
        (0, 2, EdgeType.IDENTITY),
        (1, 3, EdgeType.IDENTITY),
        (2, 3, EdgeType.IDENTITY),
        (2, 4, EdgeType.IDENTITY),
        (3, 5, EdgeType.IDENTITY)
    ]
)
```

```python
vzx.print_summary()
```

```python
qubits = { 0 : [0, 2, 4] , 1 : [1, 3, 5] }
layers = { 0 : [0,1], 1 : [2,3], 2 : [4,5] }

for qubit, nodes in qubits.items():
    for node in nodes:
        vzx.get_zx_node(node).qubit = qubit

for layer, nodes in layers.items():
    for node in nodes:
        vzx.get_zx_node(node).layer = layer
```
